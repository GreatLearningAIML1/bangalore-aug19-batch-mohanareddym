*The results from three experiments with overlapping 95% confidence intervals are given below, and we want to confirm that the results for all three experiments are not significantly different.*

But before conducting ANOVA, test equality of variances (using Levene's test) is satisfied or not. If not, then mention that we cannot depend on the result of ANOVA

```python
import numpy as np
import pandas as pd
from scipy.stats import levene


e1 = np.array([1.595440,1.419730,0.000000,0.000000])
e2 = np.array([1.433800,2.079700,0.892139,2.384740])
e3 = np.array([0.036930,0.938018,0.995956,1.006970])

#Assumption: All the 3 datasets (e1,e2 & e3) are random, independent, parametric & normally distributed
```

Perform levene test on the data

The Levene test tests the null hypothesis that all input samples are from populations with equal variances. Levene’s test is an alternative to Bartlett’s test bartlett in the case where there are significant deviations from normality.

source: scipy.org

#### Answer:

```python
### Assuming that the aplha value is 5% 
###Null Hypothesis is that all the samples have the same equal variances mu1=mu2=mu3.
###Alternate hypotheis is that there is a significance difference in variancesmu1is not equal to mu2 is not equal to mu3.

print("The lenght , mean and std of array e1 is ",len(e1),np.mean(e1),np.std(e1,ddof=1))
print("The lenght , mean and std of array e2 is ",len(e2),np.mean(e2),np.std(e2,ddof=1))
print("The lenght , mean and std of array e3 is ",len(e3),np.mean(e3),np.std(e3,ddof=1))

print(levene(e1,e2,e3))

print(" As the P value is more than 5%, there fore we failed to reject the null hypothesis which states that all the samples have the same equal variances")    
    
```

## Question 5

The one-way ANOVA tests the null hypothesis that two or more groups have the same population mean. The test is applied to samples from two or more groups, possibly with differing sizes.

use stats.f_oneway() module to perform one-way ANOVA test

```python
### Null Hypothesis is mu1=mu2=mu3.(There is no signifance difference btwn the data sets.)
### Alternate Hypothesis is atleast one of the mu value is different from the other which
                            ### states that there is a signifance differnce btwn the data sets)
import pandas as pd
import statsmodels.api         as     sm
from   statsmodels.formula.api import ols

e1 = np.array([1.595440,1.419730,0.000000,0.000000])
e2 = np.array([1.433800,2.079700,0.892139,2.384740])
e3 = np.array([0.036930,0.938018,0.995956,1.006970])

print("The lenght , mean and std of array e1 is ",len(e1),np.mean(e1),np.std(e1,ddof=1))
print("The lenght , mean and std of array e2 is ",len(e2),np.mean(e2),np.std(e2,ddof=1))
print("The lenght , mean and std of array e3 is ",len(e3),np.mean(e3),np.std(e3,ddof=1))


```

```python
array_df = pd.DataFrame()

df1            = pd.DataFrame({'e': '1', 'array_data':e1})
df2            = pd.DataFrame({'e': '2', 'array_data':e2})
df3            = pd.DataFrame({'e': '3', 'array_data':e3})

array_df = array_df.append(df1) 
array_df = array_df.append(df2) 
array_df = array_df.append(df3) 
```

```python
array_df.head()
```

```python
import statsmodels.api         as     sm
from   statsmodels.formula.api import ols

mod = ols('array_data ~ e', data = array_df).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print(aov_table)

print("As the value of P is more than the value of Aplha the statistical decision is to failed to reject the null hypothesis at Aplha is at 5%")
```

Let's create a small dataset for dice rolls of four players

```python
import numpy as np

d1 = [5, 8, 3, 8]
d2 = [9, 6, 8, 5]
d3 = [8, 12, 7, 2]
d4 = [4, 16, 7, 3]
d5 = [3, 9, 6, 5]
d6 = [7, 2, 5, 7]

dice = np.array([d1, d2, d3, d4, d5, d6])
```

run the test using SciPy Stats library

Depending on the test, we are generally looking for a threshold at either 0.05 or 0.01. Our test is significant (i.e. we reject the null hypothesis) if we get a p-value below our threshold.

For our purposes, we’ll use 0.01 as the threshold.

use stats.chi2_contingency() module 

This function computes the chi-square statistic and p-value for the hypothesis test of independence of the observed frequencies in the contingency table

Print the following:

- chi2 stat
- p-value
- degree of freedom
- contingency



```python
dice = np.array([d1, d2, d3, d4, d5, d6]) 
chi_sq_Stat, p_value, deg_freedom, exp_freq = stats.chi2_contingency(dice)
print(" The chi_sq_Stat, p_value, deg_freedom ,contigency",chi_sq_Stat, p_value, deg_freedom,exp_freq)
```

## Question 10

A Paired sample t-test compares means from the same group at different times.

The basic two sample t-test is designed for testing differences between independent groups. 
In some cases, you might be interested in testing differences between samples of the same group at different points in time. 
We can conduct a paired t-test using the scipy function stats.ttest_rel(). 

```python
 ## Creates a normal distribution with a mean value of 100 and std of 30
before= stats.norm.rvs(scale=30, loc=100, size=500)
after = before + stats.norm.rvs(scale=5, loc=-1.25, size=500)
```

Test whether a weight-loss drug works by checking the weights of the same group patients before and after treatment using above data.

```python
from scipy.stats             import ttest_1samp,ttest_ind, wilcoxon
from statsmodels.stats.power import ttest_power
import matplotlib.pyplot     as     plt

###Null hypothesis is there is no significant change in the weigths mu1-mu2=0
##Alternate hypothesis is there is a significant change in the weight mu1 is not equal to mu2 

t_statistic, p_value  =  stats.ttest_ind(before,after)
print(p_value)
print("As the pvalue is more than the value of Alpha which is 5% we therefore failed to reject the null hypothesis at this level of significance, where as here we accept the null hypothesis")
```
